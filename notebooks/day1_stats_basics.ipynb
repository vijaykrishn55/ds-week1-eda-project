```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("../data/retail_sales_dataset.csv")
```

```python
print(df.describe())
print(df.dtypes)
```

```python
print(df['Product Category'].value_counts())
print(df.groupby('Product Category')['Total Amount'].mean())
```

```python
#histogram 
plt.figure(figsize=(8,4))
plt.hist(df['Total Amount'],bins=30, color='steelblue', edgecolor='black')
plt.title('sales distribution')
plt.xlabel('sale amount')
plt.ylabel('frequency')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x='Product Category',y='Total Amount', data =df)
plt.title('sales by category')
plt.show()
```

```python
#pivot table
pivot=pd.pivot_table(df, values='Total Amount', index='Product Category', columns= 'Gender', aggfunc='sum')
print(pivot)
```
