```python
import pandas as pd

import random 
```

```python
# mony hall problem


def monty_hall(switch=True, trails=10000):
        wins =0
        for _ in range(trails):
                car = random.randint(0,2)
                choice = random.randint(0,2)

                if switch:
                        wins += (choice != car)
                else: 
                        wins +=(choice == car)

        return wins/trails

print("switch win rate:", monty_hall(switch=True))
print("stay win rate:", monty_hall(switch=False))
```

```python
df= pd.read_csv("../data/retail_sales_dataset.csv")

#probability

#p customer is female
total = len(df)
p_female = len(df[df['Gender']=='Female'])/total
print(f"p(female): {p_female:.3f}")

# p product category is beauty
p_beauty= len(df[df['Product Category']=='Beauty'])/total
print(f"p(beauty): {p_beauty:.3f}")

# p joint probability
p_female_and_beauty =len(df[(df['Gender']=='Female') & (df['Product Category']=='Beauty')])/total
print(f"p(female and beauty): {p_female_and_beauty:.3f}")

# p conditional probability
females= df[df['Gender']=='Female']
p_beauty_given_female = len(females[females['Product Category']=='Beauty'])/len(females)
print(f"p(beauty/female): {p_beauty_given_female:.3f}")

```


PROBABILITY FORMULAS
─────────────────────────────────────────

P(A)         = favorable / total

P(A and B)   = P(A) × P(B)          [if independent]

P(A | B)     = P(A and B) / P(B)    [conditional]

P(A | B)     = P(B|A) × P(A) / P(B) [Bayes Theorem]
